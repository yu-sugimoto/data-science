##### 内積

```python
# 行列の作成
np.random.seed(0)
mat1 = np.random.randint(10, size=(5, 3))
mat2 = np.random.randint(10, size=(5, 3))
mat1, mat2
```

```python
# 積
mat1 * mat2
```

```python
# 内積
np.dot(mat1, mat2.T)
```

```python
# 内積のサンプルプロダクト（各曜日の売上の総合計①）
np.random.seed(0)
sales_amount = np.random.randint(20, size=(5, 3))
sales_amount
```

```python
# 各曜日のアイテムの売上個数（各曜日の売上の総合計②）
weekly_sales = pd.DataFrame(sales_amount, index=["月", "火", "水", "木", "金"], columns=["アイテム1", "アイテム2", "アイテム3"])
weekly_sales
```

```python
# アイテムの価格（各曜日の売上の総合計③）
item_prices = np.array([10, 8, 12])
item_prices = pd.DataFrame(item_prices.reshape(1, 3), index=["価格"], columns=["アイテム1", "アイテム2", "アイテム3"])
item_prices
```

```python
# 各曜日の売上の総合計（各曜日の売上の総合計④）
daily_sales = np.dot(weekly_sales, item_prices.T)
daily_sales
```

```python
# データフレームに追加（各曜日の売上の総合計⑤）
weekly_sales["合計"] = daily_sales
weekly_sales
```
